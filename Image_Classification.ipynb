## Scanpath Images

####DP

```python
imageFolder = 'Scanpath Images - DP'
path = path 
for i in range(10):
  print(f'Run {i+1} out of 10')
  modelName = f'model_scanpath_images_dp_run_{i}'
  getValResults(imageFolder,path,modelName,positiveClass='DP',patience=10)
```

```python
meanResults = pd.DataFrame.from_dict({'test_acc':[],'test_auc':[],'negative_class_ratio':[],'positive_class_ratio':[]} )
imageFolder = 'Scanpath Images - DP'
path= path
for i in range(10):
  modelName = f'model_scanpath_images_dp_run_{i}'
  testResultsDf = getTestResults(imageFolder,path,modelName,positiveClass='DP')
  meanResults = pd.concat([meanResults,testResultsDf])

dpScanResults = meanResults.groupby(meanResults.index).agg({'test_acc':['mean','std'],'test_auc':['mean','std'],'positive_class_ratio':['mean'],'negative_class_ratio':['mean']})
```

```python
dpScanResults.to_csv(path+'/DP_Mean_Scanpath_Results_AutoEpoch.csv',index=False)
```

```python
pd.read_csv(path+'/DP_Mean_Scanpath_Results_AutoEpoch.csv').test_auc.iloc[1:].astype(float).mean()
```

# Random Hybrids

AP

```python
aucDist = []
r = 100
for i in range(r):
  print('\033[1m' + f'Step {i+1} out of {r}:' + '\033[0m')
  hybridAUC = randomHybridAUC(path,imageFolder='Scanpath Images - AP',positiveClass='AP',patience=10)
  aucDist.append(hybridAUC)
  pd.DataFrame.from_dict({'auc':aucDist}).to_csv(path+'/randomAuc100AP.csv',index=False)
```

```python
x = [list(pd.read_csv(path+'/randomAuc100AP.csv').auc.astype(float))]

# the histogram of the data
n, bins, patches = plt.hist(x, 5, density=True, facecolor='blue', alpha=0.75)

plt.xlabel('AUC Score')
plt.ylabel('Probability Density')
plt.text(0.74, 6, 'Actual AUC')
#plt.xlim(40, 160)
#plt.ylim(0, 0.03)
plt.axvline(x=0.8, color='red', linestyle='--')
plt.show()
```
