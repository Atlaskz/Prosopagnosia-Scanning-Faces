#Training Baseline Models

```python
positiveClass = ['AP','DP']
auc = {'AP':[], 'DP':[]}
fixations = {'AP':[], 'DP':[]}
k = 20

for pc in positiveClass:
    # import data from csv files 
    dfRaw = getData(path,labels=[f'{pc} - New ROI Data','Control - New ROI Data'],phases=['Learning','Target','Distractor'])
    # choose train and test datasets
    train, test = getTrainTest(path,positiveClass=f'{pc}',valSet=False,getSubjectsFrom='DataFrame',df=dfRaw)
    # train models for 1<fixations<=20
    for i in range(2,k+1):
      print(f'using {i} fixations (max={k})')
      # edit dataframe to include up to nth fixation
      allTrials = getTrialData(dfRaw,i,positive=f'{pc}')
      # process the features to make the data ready for training
      df = processData(allTrials,i)
      # train data and get test results
      result = trainData(df,train,test)
      # record the fixation number and auc for the plot
      fixations[pc].append(i)
      auc[pc].append(result)
```

# AP and Control Classification Using The Baseline Model (Logistic Regression)

```python
fig = go.Figure(data=[go.Bar(x=fixations['AP'], y=auc['AP'])])
fig.update_xaxes(title='Number of Fixations',title_font_size=30,tickfont=dict(size=25))
fig.update_yaxes(title='Performance (AUC)',title_font_size=30,tickfont=dict(size=25),range=[0.45, 0.67])
fig.update_traces(marker_color='grey')
fig.update_layout(paper_bgcolor='white',plot_bgcolor='white',width=1500)
```
